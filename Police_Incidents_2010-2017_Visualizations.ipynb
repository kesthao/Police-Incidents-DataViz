## Data Visualization

```python
# Import required libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

import pickle

%matplotlib inline
```

```python
# Load pickled dataframes
df2 = pd.read_pickle('df2.pickle')

top_3_incidents = pd.read_pickle('top_3_incidents.pickle')
df_top3_grouped = pd.read_pickle('df_top3_grouped.pickle')

top_10_incidents = pd.read_pickle('top_10_incidents.pickle')
df_top10_grouped = pd.read_pickle('df_top10_grouped.pickle')

day_of_week_count = pd.read_pickle('Day_of_Week_Count.pickle')
```

### Data Visualization #1 

```python
#------------------------
# DEFINE STYLE ATTRIBUTES
#------------------------
font = {'family': 'sans-serif',
        'color':  'black',
        'weight': 'bold',
        'size': 18,
        }

grids = {
    'color':'b0b0b0',
    'linestyle': '-',
    'linewidth': 1.0,
    'alpha': 1.0
    }

color_scheme = ['r', 'gold', 'b', 'g', 'aquamarine', 'gold', 'coral', 'sienna', 'magenta', 'navy']

SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 24

#---------------------------------------------------------------------
# PLOTTING START HERE
#---------------------------------------------------------------------
fig = plt.figure(figsize=(15, 15))
# plt.rc('figure', titlesize=TITLE_SIZE)  # fontsize of the figure title
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot
for i,c in enumerate(top_10_incidents.index.tolist()):
    ax.plot(df_top10_grouped[df_top10_grouped['Description']==c]['Day_of_Week_Num'], 
            df_top10_grouped[df_top10_grouped['Description']==c]['Count'], 'o-', 
            lw=3.0, ms=10.0, label=c, c=color_scheme[i])

#------------------------
# X-Axis
#------------------------
plt.xlabel('Day of Week', fontdict=font)
# ax.set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
# ax.xaxis.set_ticks_position('bottom')
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels

#------------------------
# Y-Axis
#------------------------
plt.ylabel('# of Incidents', fontdict=font) # y-label
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
# ax.set_yticks()
# ax.set_ylims()
# ax.yaxis.set_ticks_position('left')
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels

#------------------------
# Axes attributes
#------------------------
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels

# Spines
# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)
# ax.spines['left'].set_visible(False)
# ax.spines['bottom'].set_visible(False)

#------------------------
# Legend
#------------------------
plt.rc('legend', fontsize=SMALL_SIZE, )    # legend fontsize
ax.legend(loc=1)
for text in ax.legend().get_texts():
    plt.setp(text, color = 'black')
    
#------------------------
# Grid
#------------------------
# plt.grid(False)
# plt.rc('grid', **grids)

ax.set_title("Top 10 Incident Types (2010 - 2017)", fontdict=font)

fig.autofmt_xdate()
# plt.show()

plt.savefig('Viz1.png')
```

## Data Visualization #3

```python
#------------------------
# DEFINE STYLE ATTRIBUTES
#------------------------
sns.set_style('whitegrid')
tfont = {'family': 'sans-serif',
        'color':  '#546f8c',
        'weight': 'bold',
        'size': 40,
        }
axfont = {'family': 'sans-serif',
        'color':  'gray',
        'size': 18
        }

grids = {
    'color':'gray',
    'linestyle': '--',
    'linewidth': 1.0,
    'alpha': 0.5
    }

SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 24

#---------------------------------------------------------------------
# PLOTTING START HERE
#---------------------------------------------------------------------
fig = plt.figure(figsize=(17, 10))
# plt.rc('figure', titlesize=TITLE_SIZE)  # fontsize of the figure title
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot
ax.plot(sorted(day_of_week_count.index.tolist()), day_of_week_count.values.tolist(),
        lw=10.0, alpha=0.7, color='orange')

# Plot
for i,c in enumerate(top_3_incidents.index.tolist()):
    ax.plot(df_top3_grouped[df_top3_grouped['Description']==c]['Day_of_Week_Num'],
            df_top3_grouped[df_top3_grouped['Description']==c]['Count'],
            lw=2.0, c='grey', alpha=0.4)
    
#------------------------
# X-Axis
#------------------------
plt.xlabel('Day of Week', fontdict=axfont)
# ax.set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
ax.xaxis.set_ticks_position('bottom')
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels

#------------------------
# Y-Axis
#------------------------
plt.ylabel('# of Incidents', fontdict=axfont) # y-label
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ','))
    )
ax.yaxis.set_ticks_position('left')
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels

#------------------------
# Axes attributes
#------------------------
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels

# Spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

#------------------------
# ANNOTATION
#------------------------
plt.text(7.1, 4800, 'Other Theft (Total: 46,836)', horizontalalignment='left', color='black', fontsize=14, )
plt.text(7.1, 3600, 'Burglary Of Dwelling (Total: 27,977)', horizontalalignment='left', color='black', fontsize=14, )
plt.text(7.1, 2800, 'Theft From Motr Vehc (Total: 27,956)', horizontalalignment='left', color='black', fontsize=14, )
plt.text(7.1, 21000, 'All Incidents (Total: 169,554)', horizontalalignment='left', color='orange', fontsize=20, )

for i,d in enumerate(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']):
    plt.text(i+1, 1000, '('+d+')',color='gray',fontsize=14 )
    
#------------------------
# Grid
#------------------------
plt.grid(True)
ax.grid(ls='dashed', lw=1.0, alpha=0.5)
# plt.rc('grid', **grids)

ax.set_title("Most Police Incidents Reported on Monday", fontdict=tfont)

fig.autofmt_xdate()
# plt.show()

plt.savefig('Viz3.png')
```
